## <a id='missing'>Periksa Data yang Hilang</a>  

Mari kita buat sebuah fungsi untuk memeriksa apakah ada data yang hilang


```python
# Tulis kode di atas pada bagian kode berikut dan jalankan
def missing_data(data):
    total = data.isnull().sum().sort_values(ascending = False)
    percent = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    return pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])

missing_data(train)
```

```python
# Tulis kode di atas pada bagian kode berikut dan jalankan
missing_data(test)
```

Baik dalam dataset **train** maupun **test**, tidak ada data yang hilang.


# <a id='preprocess'>Praproses Data</a>  
Konversi data kategorikal menjadi numerik.

```python
# Tulis kode di atas pada bagian kode berikut dan jalankan
cat_columns = train.select_dtypes(['object']).columns
train[cat_columns] = train[cat_columns].astype('category')
train[cat_columns] = train[cat_columns].apply(lambda x:x.cat.codes)
```

```python
# Tulis kode di atas pada bagian kode berikut dan jalankan
cat_columns = test.select_dtypes(['object']).columns
test[cat_columns] = test[cat_columns].astype('category')
test[cat_columns] = test[cat_columns].apply(lambda x:x.cat.codes)
```

# <a id='model'>Bangun Model Prediktif</a>  

Mari kita mulai dengan pembuatan model prediktif. Kita akan membuat model yang sangat sederhana sebagai langkah awal.


## <a id='split'>Pisahkan Data</a>  

Mari kita bagi dataset menjadi set **pelatihan** (train) dan **validasi** (valid). Kita akan menggunakan pembagian **80-20**.  
Selain itu, kita juga akan menetapkan matriks untuk pelatihan dan validasi serta vektor dengan nilai target.


```python
# Tulis kode di atas pada bagian kode berikut dan jalankan

from sklearn.model_selection import train_test_split

train, valid = train_test_split(train, shuffle=True)
```

## <a id='baseline'>Buat Model Dasar (Baseline Model)</a>  


```python
# Tulis kode di atas pada bagian kode berikut dan jalankan


target = 'Weather Type'
train_Y = train[target].values
train_X = train.drop(target,axis=1)

valid_Y = valid[target].values
valid_X = valid.drop(target,axis=1)
```

Mari kita siapkan model sederhana menggunakan **Naive Bayes**.

NaÃ¯ve Bayes juga dikenal sebagai **klasifikator probabilistik** karena didasarkan pada **Teorema Bayes**.  
NaÃ¯ve Bayes adalah klasifikator berbasis probabilitas yang mengasumsikan bahwa setiap fitur bersifat independen satu sama lain (oleh karena itu disebut 'naÃ¯ve').  

Model ini menghitung probabilitas bahwa suatu data termasuk dalam kelas tertentu berdasarkan nilai fitur-fiturnya.  
NaÃ¯ve Bayes dapat diibaratkan sebagai "penebak cerdas" yang menggunakan data sebelumnya untuk memprediksi hasil di masa depan, berdasarkan prinsip:  
*"Jika diberikan kondisi ini, berapa kemungkinan terjadinya kondisi itu?"*


```python
# Tulis kode di atas pada bagian kode berikut dan jalankan

from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB(var_smoothing=1e-09)
```

Selanjutnya, kita melatih (fit) klasifikator dengan data pelatihan yang telah disiapkan sebelumnya.


```python
# Tulis kode di atas pada bagian kode berikut dan jalankan

gnb.fit(train_X, train_Y)
```

```python
# Tulis kode di atas pada bagian kode berikut dan jalankan

pred = gnb.predict(valid_X)
```


## <a id='52'>Evaluasi Model</a>  

Mari kita evaluasi kinerja model.  

Pertama, kita akan mengevaluasi **akurasi** pada dataset pelatihan.


```python
# Tulis kode di atas pada bagian kode berikut dan jalankan

gnb.score(train_X, train_Y)
acc = round(gnb.score(train_X, train_Y)* 100, 2)
print("Naive Bayes accuracy (train set):", acc)
```

Hasil ini menunjukkan bahwa jumlah total prediksi yang benar dibagi dengan jumlah total contoh dalam dataset pelatihan adalah **100%**.

Selanjutnya, kita akan mengevaluasi **akurasi** pada dataset validasi.


```python
# Tulis kode di atas pada bagian kode berikut dan jalankan
gnb.score(valid_X, valid_Y)
acc = round(gnb.score(valid_X, valid_Y) * 100, 2)
print("Naive Bayes accuracy (validation set):", acc)
```

Akurasi pada dataset pelatihan jauh lebih tinggi dibandingkan dengan akurasi pada dataset validasi.


## <a id='svm'>SVM</a>  

**Support Vector Machine (SVM)** adalah algoritma **machine learning** yang digunakan untuk **klasifikasi** dan **regresi**.  
Dalam klasifikasi, SVM bekerja dengan mencari **hyperplane optimal** yang memisahkan kelas-kelas dalam suatu dataset dengan **margin terbesar**.


```python
# Tulis kode di atas pada bagian kode berikut dan jalankan
from sklearn.svm import SVC

# Inisialisasi model SVM
svm = SVC(kernel='rbf', C=1.0, gamma='scale') #Anda dapat menyesuaikan kernel, C, dan gamma

#Latih model
svm.fit(train_X, train_Y)

#Lakukan prediksi
pred = svm.predict(valid_X)

#Hitung akurasi pada dataset pelatihan
train_acc = round(svm.score(train_X, train_Y) * 1000, 2)
print("Akurasi SVM (dataset pelatihan):", train_acc)

#Hitung akurasi pada dataset validasi
valid_acc = round(svm.score(valid_X, valid_Y) * 1000, 2)
print("Akurasi SVM (dataset validasi):", valid_acc)
```

## <a id='knn'>KNN</a>  

**K-Nearest Neighbors (KNN)** dapat diibaratkan seperti mencari teman yang paling mirip dengan Anda.  
Ketika perlu membuat keputusan (misalnya, menentukan kelas suatu data), algoritma ini melihat **'k' tetangga terdekat** di sekitarnya.  

Keputusan akhir ditentukan berdasarkan mayoritas kelas dari tetangga-tetangga tersebut.  
Sebagai contoh, jika sebagian besar tetangga terdekat Anda adalah **kucing**, maka kemungkinan besar Anda juga **kucing**.


```python
# Tulis kode di atas pada bagian kode berikut dan jalankan

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5, weights= 'uniform', metric='minkowski')
knn.fit(train_X, train_Y)
pred = knn.predict(valid_X)

knn.score(train_X, train_Y)
acc = round(knn.score(train_X, train_Y) * 100, 2)
print("KNN accuracy (train set):", acc)

knn.score(valid_X, valid_Y)
acc = round(knn.score(valid_X, valid_Y) * 100, 2)
print("KNN accuracy (validation set):", acc)
```

Akurasi pada data pelatihan jauh lebih tinggi dibandingkan dengan akurasi pada dataset validasi.  
Model terlalu **baik dalam menghafal** data pelatihan, yang berarti kita mengalami **overfitting**.
